#  Exploratory Data Analysis

This notebook is to explore the dataset, learn about its structure, size and other relevant characteristics.

```python
from src.data.loader import load_excel
import matplotlib.pyplot as plt
from constants import DataSetConfig as DataSetConfig
from IPython.display import display
import seaborn as sns
```

```python
df = load_excel(DataSetConfig.PATH, DataSetConfig.COLUMNS)
df.info()
display(df.head())
display(df.describe())
```

## Body

Description of the 'Body' feature:

```python
display(df[DataSetConfig.FEATURES_BODY].describe().to_frame().T)

df['Body_char_count'] = df[DataSetConfig.FEATURES_BODY].str.len()
display(df['Body_char_count'].describe().to_frame().T)

sns.histplot(df['Body_char_count'], bins=50)
plt.title("Distribution of Character Counts")
plt.xlabel("Number of characters")
plt.ylabel("Frequency")
plt.show()

df['Body_word_count'] = df[DataSetConfig.FEATURES_BODY].str.split().str.len()
display(df['Body_word_count'].describe().to_frame().T)

sns.histplot(df['Body_word_count'], bins=50)
plt.title("Distribution of Word Counts")
plt.xlabel("Number of words")
plt.ylabel("Frequency")
plt.show()
```

## Scores

What's the scores distribution?

```python
df['rounded_score'] = df[DataSetConfig.FEATURES_TARGET].round()

bins = [i - 0.5 for i in range(1, 9)]
plt.hist(df['rounded_score'], bins=bins, edgecolor='black')
plt.xticks(range(1, 9))
plt.xlabel("Average Score (rounded)")
plt.ylabel("Frequency")
plt.title("Histogram of Average Scores (Rounded)")
plt.show()
```

What's the scores distribution by topic?

```python
df['Topic_Number'] = df['Topic_1'].str[:2].astype(int)

topic_score_counts = df.groupby(["Topic_Number", 'rounded_score']).size().unstack(fill_value=0)

ax = topic_score_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribution of Rounded Average Scores by Topic', y=1.2)
plt.legend(title='Rounded Score', bbox_to_anchor=(0.5, 1.05), loc='lower center', ncol=len(topic_score_counts.columns))
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

What's the scores distribution by presentation type?

```python
presentation_type_score_counts = df.groupby(["Presentation_Type", 'rounded_score']).size().unstack(fill_value=0)

ax = presentation_type_score_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribution of Rounded Average Scores by Presentation Type', y=1.3)
plt.legend(title='Rounded Score', bbox_to_anchor=(0.5, 1.05), loc='lower center', ncol=len(presentation_type_score_counts.columns))
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
